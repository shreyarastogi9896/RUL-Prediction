```python
correlation_matrix = dft1.corr()

# Plot the heatmap
plt.figure(figsize=(60, 60))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```

```python
low_correlation_columns = [col for col in dft1.columns if abs(dft1[col].corr(dft1['RUL'])) < 0.5]
```

```python
dft1_new = dft1.drop(columns=low_correlation_columns)
dft1_new.info()
```

```python
X = dft1_new.iloc[:, 0:18]
Y=dft1_new.iloc[:,18]
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 2)
regr = RandomForestRegressor()
regr.fit(X_train,Y_train)
Y_pred = regr.predict(X_test)
print('RMSE on Test set:', np.sqrt(metrics.mean_squared_error(Y_test, Y_pred)))
Y_predt = regr.predict(X_train)
print('RMSE on Training set:', np.sqrt(metrics.mean_squared_error(Y_train, Y_predt)))
r2 = r2_score(Y_test, Y_pred)

print(f"R-squared: {r2:.2f}")
```

```python
X = dft1.iloc[:, 0:26]
Y=dft1.iloc[:,26]
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 2)
regr = RandomForestRegressor()
regr.fit(X_train,Y_train)
Y_pred = regr.predict(X_test)
print('RMSE on Test set:', np.sqrt(metrics.mean_squared_error(Y_test, Y_pred)))
Y_predt = regr.predict(X_train)
print('RMSE on Training set:', np.sqrt(metrics.mean_squared_error(Y_train, Y_predt)))
r2 = r2_score(Y_test, Y_pred)
print(f"R-squared: {r2:.2f}")
```

```python
from sklearn.svm import SVR
scaler = StandardScaler()
X = dft1.iloc[:, 1:26]
Y=dft1.iloc[:,26]
X_scaled = scaler.fit_transform(X)
X_train, X_test, Y_train, Y_test = train_test_split(X_scaled, Y, test_size = 0.2, random_state = 42)
svr = SVR(kernel='rbf', C=100, epsilon=5)
svr.fit(X_train, Y_train)
Y_pred = svr.predict(X_test)
rmse = np.sqrt(mean_squared_error(Y_test, Y_pred))
mae = mean_absolute_error(Y_test, Y_pred)
print(f"RMSE: {rmse:.2f}")
print(f"MAE: {mae:.2f}")
r2 = r2_score(Y_test, Y_pred)
print(f"R-squared: {r2:.2f}")
```

```python
# Hyperparameter tuning for SVR model
param_grid = {
    'C': [1, 10, 100],
    'epsilon': [0.1, 1, 5],
    'kernel': ['linear', 'rbf', 'poly']
}
svr = SVR()
random_search = RandomizedSearchCV(svr, param_grid, n_iter=10,cv=3, scoring='neg_mean_squared_error')
random_search.fit(X_train, Y_train)

# Train the best model
best_svr = random_search.best_estimator_
Y_pred = best_svr.predict(X_test)

# Evaluate model
rmse = np.sqrt(mean_squared_error(Y_test, Y_pred))
mae = mean_absolute_error(Y_test, Y_pred)
r2 = r2_score(Y_test, Y_pred)
print(f"RMSE: {rmse:.2f}")
print(f"MAE: {mae:.2f}")
print(f"R-squared: {r2:.2f}")
```

```python
xgb = XGBRegressor(objective='reg:squarederror', n_estimators=100, learning_rate=0.1, max_depth=5)
xgb.fit(X_train, Y_train)
Y_pred = xgb.predict(X_test)
rmse = np.sqrt(mean_squared_error(Y_test, Y_pred))
mae = mean_absolute_error(Y_test, Y_pred)
r2 = r2_score(Y_test, Y_pred)
print(f"RMSE: {rmse:.2f}")
print(f"MAE: {mae:.2f}")
print(f"R-squared: {r2:.2f}")
```
